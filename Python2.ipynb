```python
def bienvenida(nombre, lenguaje = 'Python'):
    print('¡Bienevenido a', lenguaje, nombre + '!')
    return
bienvenida('Jest')
```

```python
def menu(*platos):
    print('Hoy tenemos: ', end='')
    for plato in platos:
        print(plato, end=', ')
    return

menu('pasta', 'pizza', 'ensalada')
```
