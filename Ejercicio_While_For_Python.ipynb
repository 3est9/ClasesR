```python
# Bucle con while en una list()
```

```python
frutas = ["fresa", "plátano", "manzana", "pera", "cereza"]
i = 0

while i < len(frutas):
    if i != 1:
        print(frutas[i])
    i += 1
```

```python
# Bucle con for en una list()
```

```python
frutas = ["fresa", "plátano", "manzana", "pera", "cereza"]

for i in range(len(frutas)):
    if i == 1:
        continue
    print(frutas[i])
```
